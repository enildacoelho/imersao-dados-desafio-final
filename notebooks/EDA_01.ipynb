# Análise exploratória do conjunto de dados "Dados Experimentos - MoA"



```python
# Obtem um conjunto de ferramentas/bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
import seaborn as sns

# Carrega dados (formato  csv, compressão do tipo zip) a partir de uma URL 

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados_experimentos = pd.read_csv(url_dados, compression='zip')
```
